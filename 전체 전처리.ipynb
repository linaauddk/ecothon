### 데이터 형식 표준화


```python

def standardize(df, year):
    df = df.copy()
    
    if year in [2020, 2021, 2022]:
        df.rename(columns={
            '등록일자': '측정시간',
            '기온(℃)': '온도 평균(℃)',
            '상대습도( %)': '습도 평균(%)'
        }, inplace=True)
        df['측정시간'] = pd.to_datetime(df['측정시간'], errors='coerce')

    elif year in [2023, 2024]:
        df.rename(columns={
            '온도 평균(℃)': '온도 평균(℃)',
            '습도 평균(%)': '습도 평균(%)'
        }, inplace=True)
        # 여기에 날짜 형식 지정
        df['측정시간'] = pd.to_datetime(df['측정시간'], format="%Y-%m-%d_%H:%M:%S", errors='coerce')

    df['온도 평균(℃)'] = pd.to_numeric(df['온도 평균(℃)'], errors='coerce')
    df['습도 평균(%)'] = pd.to_numeric(df['습도 평균(%)'], errors='coerce')

    # 측정시간 기준 오름차순 정렬
    df = df.sort_values(by='측정시간', ascending=True)
    
    return df[['시리얼', '측정시간', '온도 평균(℃)', '습도 평균(%)']]
```

```python
df_2020_std = standardize(df_2020, 2020)
df_2021_std = standardize(df_2021, 2021)
df_2022_std = standardize(df_2022, 2022)
df_2023_std = standardize(df_2023, 2023)
df_2024_std = standardize(df_2024, 2024)
```

### 결측치 계산

```python
def analyze_serial_null_ratio(df, sort_desc=True):
    """
    시리얼별 온도/습도 결측률 계산 함수.
    
    Parameters:
        df: 정제된 DataFrame (ex: df_2024_std)
        sort_desc: 결측률 정렬 방향 (기본값은 내림차순)
        
    Returns:
        결측률 정보가 담긴 DataFrame
    """
    # 총 행 수
    total_counts = df.groupby('시리얼').size()
    
    # 컬럼별 결측 수
    temp_nulls = df['온도 평균(℃)'].isnull().groupby(df['시리얼']).sum()
    humid_nulls = df['습도 평균(%)'].isnull().groupby(df['시리얼']).sum()
    
    # 데이터프레임 구성
    result = pd.DataFrame({
        '온도 평균(℃)': temp_nulls,
        '습도 평균(%)': humid_nulls,
        '총 결측 수': temp_nulls + humid_nulls,
        '총 행 수': total_counts
    })
    
    # 결측률 계산 (온도/습도 두 컬럼 기준)
    result['결측률 (%)'] = (result['총 결측 수'] / (result['총 행 수'] * 2)) * 100
    result = result.fillna(0)
    
    return result.sort_values(by='결측률 (%)', ascending=not sort_desc)
```

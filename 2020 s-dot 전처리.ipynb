```python
# 온도/습도 평균이 모두 null인 시리얼별 합산
null_only_serials = (
    merged_2020.groupby("시리얼")[['기온(℃)', '상대습도( %)']]
    .apply(lambda x: x.isnull().all().all())
)

# True인 시리얼만 필터링
null_only_serials = null_only_serials[null_only_serials].index.tolist()
```

```python
null_only_serials
```

```python
# 널값만 있는 시리얼 제거
cleaned_df_2020 = merged_2020[~merged_2020['시리얼'].isin(null_only_serials)]
cleaned_df_2020
```

```python
cleaned_df_2020.isnull().sum()
```

```python
# 1. 우선 문자열을 표준화 (정규화)
cleaned_df_2020['등록일자'] = cleaned_df_2020['등록일자'].astype(str).str.replace('.', '-', regex=False).str.strip()

# 2. to_datetime으로 유연하게 처리
cleaned_df_2020['날짜'] = pd.to_datetime(cleaned_df_2020['등록일자'], errors='coerce', dayfirst=False).dt.date
```

```python
# 수치형으로 변환
cleaned_df_2020['기온(℃)'] = pd.to_numeric(cleaned_df_2020['기온(℃)'], errors='coerce')
cleaned_df_2020['상대습도( %)'] = pd.to_numeric(cleaned_df_2020['상대습도( %)'], errors='coerce')
```

```python
# 2. 그룹 평균 계산
group_avg = cleaned_df_2020.groupby(['시리얼', '날짜'])[['기온(℃)','상대습도( %)']].transform('mean')

group_avg
```

```python
# 3. 결측값에 그룹 평균으로 채우기
cleaned_df_2020['기온(℃)'] = cleaned_df_2020['기온(℃)'].fillna(group_avg['기온(℃)'])
cleaned_df_2020['상대습도( %)'] = cleaned_df_2020['상대습도( %)'].fillna(group_avg['상대습도( %)'])

# 필요 없다면 날짜 컬럼 삭제
cleaned_df_2020.drop(columns='날짜', inplace=True)
```

```python
group_avg.isnull().sum()
```

```python
cleaned_df_2020.isnull().sum()
```

```python
null_rows_2020 =cleaned_df_2020[cleaned_df_2020.isnull().any(axis=1)]
null_rows_2020
```

```python
null_by_serial = (
    null_rows_2020
    .groupby("시리얼")[['기온(℃)', '상대습도( %)']]
    .apply(lambda x: x.isnull().sum())
)

# 총 결측 수 계산
null_by_serial['총 결측 수'] = null_by_serial.sum(axis=1)

# 정렬
null_by_serial = null_by_serial.sort_values(by='총 결측 수', ascending=False)
null_by_serial
```

```python
# 시리얼별 전체 측정 횟수
total_count_by_serial = cleaned_df_2020['시리얼'].value_counts().sort_index()

# 결측률 (%) 계산
partial_nulls = null_by_serial.copy()
partial_nulls['총 행 수'] = total_count_by_serial* 2
partial_nulls['결측률 (%)'] = (partial_nulls['총 결측 수'] / (partial_nulls['총 행 수'])) * 100
```

```python
partial_nulls
```

```python
serial_null_2020=partial_nulls.sort_values("결측률 (%)",ascending=False)
serial_null_2020
```

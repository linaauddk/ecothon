```python
import os
import pandas as pd
```

```python
folder_path = "s-dot2024"
csv_files = sorted([f for f in os.listdir(folder_path) if f.endswith('.csv')])

# 필요한 컬럼만 선택
required_cols = ['시리얼', '측정시간', '자치구', '행정동', '온도 평균(℃)', '습도 평균(%)']

# 병합을 위한 빈 DataFrame 초기화
merged_df = pd.DataFrame(columns=required_cols)

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    try:
        header_check = pd.read_csv(file_path, encoding='cp949',index_col=False, nrows=1)
        missing_cols = list(set(required_cols) - set(header_check.columns))

        if missing_cols:
            print(f"{file} → 필요한 열 누락: {missing_cols}, 건너뜀")
            continue

        temp_df = pd.read_csv(file_path, encoding='cp949', usecols=required_cols,index_col=False, low_memory=False)
        
        # 수치형 컬럼을 명시적으로 변환
        temp_df['온도 평균(℃)'] = pd.to_numeric(temp_df['온도 평균(℃)'], errors='coerce')
        temp_df['습도 평균(%)'] = pd.to_numeric(temp_df['습도 평균(%)'], errors='coerce')
        temp_df['시리얼'] = temp_df['시리얼'].astype(str)
        temp_df['측정시간'] = temp_df['측정시간'].astype(str)
        temp_df['자치구'] = temp_df['자치구'].astype(str)
        temp_df['행정동'] = temp_df['행정동'].astype(str)

        
        merged_df = pd.concat([merged_df, temp_df], ignore_index=True)
        print(f"{file} → 병합 완료")

    except Exception as e:
        print(f"{file} 처리 중 오류 발생: {e}")

# 최종 Parquet로 저장
merged_df.to_parquet("merged_2024.parquet", index=False)
print("Parquet 저장 완료: merged_2024.parquet")
```

```python
# 온도/습도 평균이 모두 null인 시리얼별 합산
null_only_serials = (
    merged_2024.groupby("시리얼")[['온도 평균(℃)', '습도 평균(%)']]
    .apply(lambda x: x.isnull().all().all())
)

# True인 시리얼만 필터링
null_only_serials = null_only_serials[null_only_serials].index.tolist()
```

```python
null_only_serials
```

```python
# 널값만 있는 시리얼 제거
cleaned_df_2024 = merged_2024[~merged_2024['시리얼'].isin(null_only_serials)]
cleaned_df_2024
```

```python
cleaned_df_2024.isnull().sum()
```

```python
# 1. 날짜만 추출 (측정시간: yyyy-mm-dd HH:MM:SS → yyyy-mm-dd)
cleaned_df_2024['날짜'] = pd.to_datetime(cleaned_df_2024['측정시간'], format="%Y-%m-%d_%H:%M:%S").dt.date

cleaned_df_2024
```

```python
# 2. 그룹 평균 계산
group_avg = cleaned_df_2024.groupby(['시리얼', '날짜'])[['온도 평균(℃)', '습도 평균(%)']].transform('mean')

group_avg
```

```python
# 3. 결측값에 그룹 평균으로 채우기
cleaned_df_2024['온도 평균(℃)'] = cleaned_df_2024['온도 평균(℃)'].fillna(group_avg['온도 평균(℃)'])
cleaned_df_2024['습도 평균(%)'] = cleaned_df_2024['습도 평균(%)'].fillna(group_avg['습도 평균(%)'])

# 필요 없다면 날짜 컬럼 삭제
cleaned_df_2024.drop(columns='날짜', inplace=True)
```

```python
null_rows_2024 =cleaned_df_2024[cleaned_df_2024.isnull().any(axis=1)]
null_rows_2024
```

```python
null_by_serial = (
    null_rows_2024
    .groupby("시리얼")[['온도 평균(℃)', '습도 평균(%)']]
    .apply(lambda x: x.isnull().sum())
)

# 총 결측 수 계산
null_by_serial['총 결측 수'] = null_by_serial.sum(axis=1)

# 정렬
null_by_serial = null_by_serial.sort_values(by='총 결측 수', ascending=False)
null_by_serial
```

```python
# 시리얼별 전체 측정 횟수
total_count_by_serial = cleaned_df_2024['시리얼'].value_counts().sort_index()

# 결측률 (%) 계산
partial_nulls = null_by_serial.copy()
partial_nulls['총 행 수'] = total_count_by_serial* 2
partial_nulls['결측률 (%)'] = (partial_nulls['총 결측 수'] / (partial_nulls['총 행 수'])) * 100
```

```python
partial_nulls
```

```python
serial_null_2024=partial_nulls.sort_values("결측률 (%)",ascending=False)
serial_null_2024
```
